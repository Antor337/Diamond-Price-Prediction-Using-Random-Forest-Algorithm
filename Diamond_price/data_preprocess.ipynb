```python
import pandas as pd
```

```python
df=pd.read_csv("dataset/diamonds.csv")
```

```python
df
```

```python
quality=['Fair','Good','Very Good','Premium','Ideal']
clr=['J','I','H','G','F','E','D']
clrty=['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
def quality_code(cut):
    for i in range(len(quality)):
        if(cut==quality[i]):
            return i;

def color_code(color):
    for i in range(len(clr)):
        if(color==clr[i]):
            return i;

def clarity_code(clarity):
    for i in range(len(clrty)):
        if(clarity==clrty[i]):
            return i;
            
```

```python
df['cut']=df['cut'].apply(quality_code)
df['color']=df['color'].apply(color_code)
df['clarity']=df['clarity'].apply(clarity_code)
```
